# Modeling

Implementing Kmeans algorithm to find people with the same interests.

```python
# Import libraries
from sklearn.cluster import KMeans
```

```python
# Training and clustering function
def cluster (n_clust, data, group = 1, person_data=None):
    # Kmeans module
    kmeans = KMeans(n_clusters=n_clust, random_state=0)
    # fit data
    kmeans.fit(data)
    
    # if we want to form the teams
    if (group == 0):
        predicted = kmeans.predict(person_data)
    
    # if we want to allocate a person
    else:
        predicted = kmeans.predict(data)
        
    return predicted
    
```

```python
data_ineterests = pd.read_csv("data/interests.txt")
```

```python
data_ineterests.head()
```

```python
cluster_labels = cluster(3, data_ineterests, group=1)
```
